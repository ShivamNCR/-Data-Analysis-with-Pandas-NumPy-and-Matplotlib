```python
data1=pd.DataFrame({
    'key1':[9,2,3,4,2],
    'key2':[7,2,3,6,2],
    'key3':[3,2,3,9,2]
})
```

```python
## retrieving inner data
data2=pd.DataFrame({
    'key1':[11,2,34,4,2],
    'key2':[7,11,3,6,2],
    'key3':[0,2,3,9,5]
})
```

```python
pd.merge(data1,data2)
```

```python
## retrieving left data
pd.merge(data1,data2,how='left')
```

```python
## retrieving right data
pd.merge(data1,data2,how='right')
```
